```python
from sklearn import svm
from sklearn.datasets import make_classification
import numpy as np
x,y=make_classification(n_samples=500,n_features=52,
                        n_redundant=22
                       ,n_informative=2,
                        random_state=25)
clf=svm.SVC(
    kernel='linear',
random_state=42
)

clf.fit(x,y)
support_vectors=clf.support_vectors_
print(f"Number of support vector:{len(support_vectors)}")

prediction=clf.predict(x)

decision_values=clf.decision_function(x)
print(f"decision value range:[{decision_values.min():.2f},{decision_values.max():.2f}]")
```
