```python
soil_moisture=[56,45,67,80]

print("soil moisture levels in different fields:")
for moisture in soil_moisture:
    print(moisture,"%")
```

```python
print("\n fields that need irrigation:")
for moisture in soil_moisture:
  if moisture<50:
    print(moisture,"%-irrigation needed")
  else:
      print(moisture,"%-sufficient moisture")
```

```python
print("\n total fields checked :",len(soil_moisture))
```

```python
soil_moisture.append(90)
print("after adding soil moisture:")
print(soil_moisture)
```

```python
soil_moisture.remove(45)
print("after removing soil moisture:")
print(soil_moisture)
```

```python
soil_moisture.sort()
print("after sorting soil moisture:")
print(soil_moisture)
```

```python
print("minimum moisture:",min(soil_moisture),"%")
print("maximum moisture:",max(soil_moisture),"%")
```
